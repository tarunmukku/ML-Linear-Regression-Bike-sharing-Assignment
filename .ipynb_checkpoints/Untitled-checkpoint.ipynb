## 2. Data preparation and cleaning

### drop redundant columns
After understanding the dataset and going through the Dataset characteristics dictonary file
below columns are being dropped

```python
#dropping the reduntant column 'instant' which has recorded index values
df.drop(['instant'],axis=1,inplace=True)

#dropping dteday as year month and day are captured in sepeate columns
df.drop(['dteday'],axis=1,inplace=True)

#dropping temp, as atemp is computed using temp and hum,windspeed 
#having atemp gives more relevent information as it also the acutal temp the user feels. 
df.drop(['temp'],axis=1,inplace=True)

#dropping  casual and registered users as captured in cnt columns 
df.drop(['casual','registered'],axis=1,inplace=True)
```

### Data prepration 

```python
# converting weekday column
import calendar
# Defining the map function
def numdaytostrday(x):
    return calendar.day_name[x]

# Applying the function to weekday 
df.weekday = df.weekday.apply(numdaytostrday)
```

```python
# converting month column

# Defining the map function
def nummnthtostrmnth(x):
    return calendar.month_name[x]

# Applying the function to weekday 
df.mnth = df.mnth.apply(nummnthtostrmnth)
```

```python
#converting season column values to  into categorical string values
df.season.replace({1:"spring",2:"summer",3:"fall",4:"winter"},inplace=True)
```

```python
#converting weathersit column values to  into categorical string values
df.weathersit.replace({1:"Clear_Few_Clouds",2:"Mist_clouds",3:"Light_rain_snow_Thunderstorm",4:'Heavy_Rain_Ice Pallets_Thunderstorm_Mist'},inplace=True)
```

```python
#converting season column values to  into categorical string values
df.season.replace({1:"spring",2:"summer",3:"fall",4:"winter"},inplace=True)
```

### Creating dummy valiables 


```python
# Get the dummy variables for the feature 'mnth' and store it in a new variable - 'mnth_variables'
mnth_variables = pd.get_dummies(df['mnth'],drop_first=True)

# Get the dummy variables for the feature 'weekday' and store it in a new variable - 'weekday_variables'
weekday_variables = pd.get_dummies(df['weekday'],drop_first=True)

# Get the dummy variables for the feature 'weathersit' and store it in a new variable - 'weathersit_variables'
weathersit_variables = pd.get_dummies(df['weathersit'],drop_first=True)

# Get the dummy variables for the feature 'weathersit' and store it in a new variable - 'weathersit_variables'
season_variables = pd.get_dummies(df['season'],drop_first=True)
```

```python
#concat dummy variabes to df
df=pd.concat([df,mnth_variables],axis=1)
df=pd.concat([df,weekday_variables],axis=1)
df=pd.concat([df,weathersit_variables],axis=1)
df=pd.concat([df,season_variables],axis=1)
df.info()             
```

```python
# Drop below columns as we have created the dummies for it

df.drop(['mnth','weekday','season','weathersit'], axis = 1, inplace = True)
```

```python
df.head()
```

## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split
#splitting dataset to train and test data we are taking 70/30 ratio 
np.random.seed(0)
df_train, df_test = train_test_split(df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

### Rescaling the Features 

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars=['atemp','hum','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
# Dividing into X and Y sets for the model building
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
#using Recursive feature elimination for model building
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 20)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
X_train.columns[~rfe.support_]
X_train_rfe = X_train[col]
```

### Model 1

```python
#Analysis of detailed stastics using statsmodel
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit() 
print(lm.summary())
```

```python
X_train_new = X_train_rfe.drop(['const'], axis=1)
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
